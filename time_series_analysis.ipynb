```python
import pandas as pd
import numpy as np
from scipy.signal import argrelextrema

from datetime import datetime
```

```python
def convert_datetime(date_str):
    try:
        return datetime.strptime(date_str, "%d/%m/%y %H:%M")
    except:
        return datetime.strptime(date_str, "%d/%m/%Y %H:%M")

def convert_time(time_str):
    a, b = list(map(int, time_str.split(":")))
    return a *60 +  b
```

```python
df = pd.read_csv(r'C:\Users\Lenovo\Desktop\Data_Science\Inceptez\SOURCE\CODE_Basics\Raj_Test.txt')
#df = pd.read_csv("data.csv")
df["COMPLETE_TIME"] = df["COMPLETETIME"].apply(convert_datetime)
df["COMPTIME_MIN"] = df["CompTime"].apply(convert_time)
```

```python
dfs = [x for _, x in df.groupby('REPORT')]
```

```python
dfs[0]
```

```python
odf = dfs[0].sort_values(by=['COMPLETE_TIME'])
```

```python
a = argrelextrema(odf['COMPLETE_TIME'].values, np.less_equal, order=1)[0]
b = argrelextrema(odf['COMPLETE_TIME'].values, np.greater_equal, order=1)[0]
```

```python
b
```

```python
spikes = []
drops = []

for odf in dfs:
    odf = odf.sort_values(by=['COMPLETE_TIME'])
    
    a = argrelextrema(odf['COMPTIME_MIN'].values, np.less_equal, order=3)[0]
    b = argrelextrema(odf['COMPTIME_MIN'].values, np.greater_equal, order=3)[0]
    
    drops.append(odf.iloc[a])
    spikes.append(odf.iloc[b])
```
